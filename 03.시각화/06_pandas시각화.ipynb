```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd  
import seaborn as sns
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
# %matplotlib inline  주피터 노느북에서 그래프를 볼 때 사용하는 매직명령어
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

```python
iris.plot()
```

#예시

```python
np.random.seed(2021)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2021', periods=100),
                   columns=['A', 'B', 'C']).cumsum() #큐블렛 섬 (조금씩더해나감)
df1.tail()
```

```python
df1.plot()
plt.title("Pandas의 Plot메소드 사용 예")
plt.xlabel("시간")
plt.ylabel("Data")
plt.show()
```

##Bar chart

```python
iris.sepal_length[:20].plot(kind='barh', rot=0, color=plt.get_cmap('hot')(np.linspace(0.8,0.2,20))) #컬러지정 추가
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].plot(kind='barh', rot=0, color=plt.get_cmap('CMRmap')(np.linspace(0.8,0.2,20))) #컬러지정 추가
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")
plt.xlabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].sort_values().plot(kind='barh', rot=0, color=plt.get_cmap('CMRmap')(np.linspace(0.8,0.2,20)))  
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")
plt.xlabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.ylim(0, 8)
plt.show()
```

```python
df2.T
```

```python
df2.T.plot.bar(rot=0)
plt.title('각 feature의 종별 평균')
plt.show()
```

```python
df3 = titanic['class'].value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객비율')
plt.axis('equal')
plt.show()
```

##Histogram

```python
iris.plot.hist()
plt.title("각 Feature 값들의 빈도수 Histogram")
plt.xlabel("데이터 값")
plt.show()
```

##Box plot

```python
iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box Plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
iris.boxplot(by='species') #박스플롯을 그룹으로
 
plt.tight_layout(pad=3, h_pad=1)
plt.title("각 Feature의 종 별 데이터에 대한 Box Plot")
plt.show()
```
