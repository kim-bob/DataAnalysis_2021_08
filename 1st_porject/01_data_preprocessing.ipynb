```python
df1 = df
```

```python
df1['월'] = df1['요양개시일자'].astype(str).str.slice(start=4, stop=6)
```

```python
df['분기'] = df['월'].astype(int).apply(lambda x: 4 if x>=10 else 3 if x>=7 else 2 if x>=4 else 1)
```

```python
df1['주상병코드'] = df1['주상병코드'].astype(str).str.slice(start=0, stop=1)
# df['주상병코드'] = df['주상병코드'].apply(lambda x : x[:3] if len(x)>=3 else x)
```

```python
df1.head()
```

```python
df2=df1
```

```python
df3 = df2.groupby('가입자 일련번호')['시도코드'].agg(**{'시도코드': lambda x : x.mode()[0]}).reset_index()
```

```python
df3.rename(columns={'시도코드':'시도코드2'}, inplace=True) 
```

```python
df4 = pd.merge(df2, df3, on=['가입자 일련번호'], how='left')
```

```python
df4.head()
```

```python
def ageagg(x):
    if x >= 17:
        return 38
    elif x>= 15:
         return 37
    elif x>=13:
        return 36
    elif x>=11:
        return 35
    elif x>=9:
        return 34
    elif x>=7:
        return 33
    elif x>=5:
        return 32
    elif x>=3:
        return 31
    else:
        return 30
```

```python
df4['연령대코드'] =  df4.연령대코드.apply(ageagg)
```
