```python
from functools import wraps

def para(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        return "<p>"+str(func(*args, **kwargs))+"<p>"
    return wrapper

@para
def outname(name):
    return  'Name :' + name

@para
def outage(age):
    return 'Age : ' + age

```

```python
print(outname('Jamses Bond'))
print(outname.__name__)

print(outage('29'))
print(outage.__name__)
```

```python
def plus(x, y):
    return x+y
def minus(x, y):
    return x-y
def multiply(x, y):
    return x*y
def divide(x, y):
    return x/y
```

```python
def dispatch_if(operator, x, y):
    if operator == 'add':
        return plus(x, y)
    elif operator == 'minus':
        return minus(x, y)
    elif operator == 'mul':
        return multiply(x, y)
    elif operator == 'div':
        return divide(x, y) 


```

```python
func_dict = {
    'add' : plus,
    'sub' :minus,
    'mul' : multiply,
    'div' : divide
}

def default_op(x,y):
    return None

def dispatch_dict(operator, x, y):
    return func_dict.get(operator, default_op)(x, y)
```
